### 2.	Remove the name and ethnicity column (to ensure confidentiality)

```python
mathdf = mathdf.drop(columns = ['Name', 'Ethinicity'], axis = 1)
phydf = phydf.drop(columns = ['Name', 'Ethinicity'], axis = 1)
dsdf = datastrudf.drop(columns = ['Name', 'Ethinicity'], axis = 1) 
```

### 3.	Fill missing score data with zero

```python
print(mathdf.isnull().sum())
print(phydf.isnull().sum())
print(dsdf.isnull().sum())
```

```python
mathdf['Score'] = mathdf['Score'].fillna(0)
print(mathdf.isnull().sum())
```

```python
phydf['Score'] = phydf['Score'].fillna(0)
print(phydf.isnull().sum())
```

```python
dsdf['Score'] = dsdf['Score'].fillna(0)
print(dsdf.isnull().sum())
```

### 4.	Merge the three files

```python
all_df = [mathdf, phydf, dsdf]
all_df
```

### 5.	Change Sex(M/F) Column to 1/2 for further analysis

```python
for dataset in all_df:
    dataset['Sex'] = dataset['Sex'].map({'M' : 1, 'F' : 2})
```

```python
all_df = pd.concat(all_df)
all_df
```
