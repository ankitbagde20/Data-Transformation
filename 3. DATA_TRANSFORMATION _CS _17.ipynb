### 4.5: Give error bar if y_error = [0.12, 0.13, 0.2, 0.1]



```python
#define data
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

#error data
y_error = [0.12, 0.13, 0.2, 0.1]

#plot data
plt.plot(x, y, linestyle = 'dashed', marker = 'o', color = 'green')

#plot on errorbars
plt.errorbar(x, y, yerr = y_error, linestyle = 'None', marker = 'None', color = 'r')

#configure  X axes
plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])

#configure  Y axes
plt.ylim(19.8,21.2)
plt.yticks([20, 21, 20.5, 20.8])

#labels
plt.xlabel("this is X")
plt.ylabel("this is Y")

#title
plt.title("Simple plot")

#show plot
plt.show()

```

### 4.6: define width, height as figsize=(4,5) DPI and adjust plot dpi=100

```python
#define plot size in inches(width, hight) & resolution(DPI)
fig = plt.figure(figsize = (4, 5), dpi = 100)

#define data
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

#error data
y_error = [0.12, 0.13, 0.2, 0.1]

#plot data
plt.plot(x, y, linestyle = 'dashed', marker = 'o', color = 'green')

#plot on errorbars
plt.errorbar(x, y, yerr = y_error, linestyle = 'None', marker = 'None', color = 'r')

#configure  X axes
plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])

#configure  Y axes
plt.ylim(19.8,21.2)
plt.yticks([20, 21, 20.5, 20.8])

#labels
plt.xlabel("this is X")
plt.ylabel("this is Y")

#title
plt.title("Simple plot")

#adjust plot
plt.subplots_adjust(left=0.18)

#show plot
plt.show()
```
