## 1.Plot Total Sales Per Month for Year 2011. How the total sales have increased over months in Year 2011. Which month has lowest Sales?

### Import Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
```

### Load data

```python
os.chdir('C:\\Users\\ankit\\OneDrive\\Documents\\Desktop\\data science\\Top Mentor Lecture\\24_Day_31_7_21\\Case Studies with Answer\\CS 18 Ans -Data Transformation-4')
df = pd.read_csv('BigMartSalesData.csv')
df.head()
```

```python
df.shape
```

### Data check

```python
df.info()
```

### Sale for year 2011 by month

```python
#get Sale for year 2011
print('Getting Sales data for year 2011')
sales_2011 = df[df['Year'] == 2011]
sales_2011
```

```python
print('Getting amount for each month')
sales_2011_month = sales_2011.groupby('Month').sum()['Amount']
sales_2011_month
```

### Plot Graph

```python
#plot the the sale Data for 2011, X Axis Sales for each month, y Axis month number

plt.plot(sales_2011_month.index, sales_2011_month.values)
plt.xlabel('sales for each month')
plt.ylabel('months number')
plt.title('sale per month in 2011')
plt.show()
```

```python
#save the plot locally
plt.savefig('Year2011MonthWiseSales')
```

## 2.Plot Total Sales Per Month for Year 2011 as Bar Chart. Is Bar Chart Better to visualize than Simple Plot?

```python
plt.bar(sales_2011_month.index, sales_2011_month.values, color = "r")
plt.xlabel('Months')
plt.ylabel('sale')
plt.title('sale per month in 201z1')
plt.show()
```

## 3.Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?


```python
sales_2011.head()
```

```python
country_sales = sales_2011.groupby(['Country']).sum()['Amount']
country_sales
```

```python
plt.title('Country wise sale in 2011')
plt.pie(country_sales.values, labels = country_sales.index, autopct = '%1.1f%%')
plt.show()
```

### 4.Plot Scatter Plot for the invoice amounts and see the concentration of amount. In which range most of the invoice amounts are concentrated

```python
invoice_sales = sales_2011.groupby(['InvoiceNo']).sum()['Amount']
invoice_sales
```

### Creat scatter plot

```python
plt.scatter(invoice_sales.values, invoice_sales.values)
plt.grid(True)
plt.show()
```
